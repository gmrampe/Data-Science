# Matplotlib

Matplotlib is a Python library used for data visualization. It provides a variety of plotting options and customization features, making it a popular choice for creating publication-quality plots.

Some common types of plots that can be created using Matplotlib include line plots, scatter plots, bar plots, histograms, and box plots, among others. Matplotlib also provides the ability to create subplots, add titles and labels to plots, change colors and line styles, and adjust other formatting options.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
#plt.scatter(x,y) is a Matplotlib function used to create a scatter plot, where x and y are arrays of data representing the x and y coordinates of each point in the plot.

#A scatter plot is a type of plot used to visualize the relationship between two numerical variables. Each point in the plot represents the value of the two variables for a single observation in the dataset. Scatter plots are useful for identifying patterns or trends in the data and for detecting any outliers or anomalies.
```

```python
plt.scatter(x,y)
plt.xlabel('this is my x axis')
plt.ylabel('this is my y axis')
plt.title('this is my vs y')
```

```python
plt.scatter(x,y, c='r')## here c respresrntes as color we required
plt.xlabel('this is my x axis')
plt.ylabel('this is my y axis')
plt.title('this is my vs y')
```

```python
plt.scatter(x,y, c='#FFA500')## c is the colour representation from html coloor codes name
plt.xlabel('this is my x axis')
plt.ylabel('this is my y axis')
plt.title('this is my vs y')
#In Matplotlib, xlabel, ylabel, and title are functions used to add labels and titles to plots.
#xlabel() is used to add a label to the x-axis of a plot, 
#while ylabel() is used to add a label to the y-axis. 
#title() is used to add a title to the plot.
```

```python
plt.scatter(x,y, c='#FFA500',alpha=.9)#The alpha parameter specifies the transparency of the markers in the plot, where a value of 0 is completely transparent and a value of 1 is completely opaque. In this case, the transparency is set to 0.9, making the markers slightly transparent.
plt.xlabel('this is my x axis')
plt.ylabel('this is my y axis')
plt.title('this is my vs y')
```

```python
plt.scatter(x,y, c='#FFA500',alpha=.5)
plt.xlabel('this is my x axis')
plt.ylabel('this is my y axis')
plt.title('this is my vs y')
```

```python
plt.scatter(x,y, c='#FFA500',alpha=.9,marker='*')
plt.xlabel('this is my x axis')
plt.ylabel('this is my y axis')
plt.title('this is my vs y')
#In Matplotlib, a marker is a symbol used to represent each data point in a plot. Markers can be used in various types of plots such as scatter plots, line plots, and bar plots, to differentiate between data points.
#For example, in a scatter plot, each marker represents a single data point. The default marker for a scatter plot is a circle, but you can change it to any of the available marker styles in Matplotlib. Some common marker styles include circles, squares, triangles, and diamonds, each with their own variations and properties.
```

```python
plt.scatter(x,y, c='#FF00FF',alpha=.9,marker='+')
plt.xlabel('this is my x axis')
plt.ylabel('this is my y axis')
plt.title('this is my vs y')
```

```python
plt.scatter(x,y, c='#FF00FF',alpha=.9,marker='+')
plt.xlabel('this is my x axis')
plt.ylabel('this is my y axis')
plt.title('this is my vs y')
plt.grid()
```

```python
#In Matplotlib, plt.figure(figsize=(width, height)) is used to create a new figure object with a specified figure size. The figsize parameter is a tuple that specifies the width and height of the figure in inches.
plt.figure(figsize=(6,2))
plt.scatter(x,y, c='#FF00FF',alpha=.9,marker='X')
plt.xlabel('this is my x axis')
plt.ylabel('this is my y axis')
plt.title('this is my vs y')
plt.grid()
```

```python
x=np.linspace(1,10, 100)
y=np.sin(x)
#In this code, np.linspace(1, 10, 100) creates a NumPy array x containing 100 evenly spaced values between 1 and 10, inclusive. np.sin(x) then applies the sine function element-wise to x to create another NumPy array y containing the sine of each value in x.
```

```python
x
```

```python
y
```

```python
#In Matplotlib, plt.plot(x, y) is a function that creates a line plot of the values in the x and y arrays. The x array specifies the x-axis values, and the y array specifies the corresponding y-axis values.
plt.plot(x,y)
```

```python
plt.scatter(x,y)
#plt.scatter is a Matplotlib function used to create a scatter plot. Scatter plots are useful for visualizing relationships between two variables, where each point in the plot represents a data point.
```

```python
plt.plot(x,y, '--r')
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y,'--y')
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y ,'--b')
plt.xlabel('this is my x data')
plt.ylabel('this is my y data')
plt.title('this is my sine wave graph')
plt.grid()
```

```python
data=[1,2,3,3,4,4,5,6,6,7,5,6,8,8,9,9]
```

```python
plt.figure(figsize=(6,3))
plt.hist(data)
plt.xlabel('represent unique value')
plt.ylabel('represent the frequency')
plt.title('histgram')
plt.grid()
```

```python
plt.figure(figsize=(6,3))
plt.hist(data, color='#A52A2A')
plt.xlabel('represent unique value')
plt.ylabel('represent the frequency')
plt.title('histgram')
plt.grid()
plt.savefig('hist.png')
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z,c='y',alpha=.8)
plt.show()
```
