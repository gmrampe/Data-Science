Bokeh is a Python library for creating interactive data visualizations in web browsers. It is built to enable the creation of complex visualizations with very large and streaming datasets. Bokeh provides a range of tools and options for creating interactive plots, charts, and graphs with ease.

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
#These lines of code import the bokeh.io and bokeh.plotting modules. The bokeh.io module provides functions for working with Bokeh output, such as displaying plots in notebooks or saving plots to files. The bokeh.plotting module provides a high-level interface for creating Bokeh plots.
#This line of code configures the Bokeh output to display plots inline in a Jupyter notebook. The output_notebook() function enables Bokeh's notebook output mode, which causes Bokeh plots to be displayed directly in the notebook. If you were working in a standalone Python script or outside of a notebook environment, you could use the output_file() function instead to specify an output file to save the plot to.
```

```python
from bokeh.plotting import figure,output_file,show
#This line of code imports the figure, output_file, and show functions from the bokeh.plotting module. The figure function is used to create a new plot, output_file is used to specify the output file for the plot, and show is used to display the plot in the default web browser.
from bokeh.sampledata.iris import flowers
```

```python
flowers
```

```python
output_file('test.html')

p=figure(title='test flower')
p.xaxis.axis_label="petal_length"
p.yaxis.axis_label="petal_width"
p.circle(flowers['petal_length'],flowers['petal_width'])
show(p)
```

```python
output_file('test.html')

p=figure(title='test flower')
p.xaxis.axis_label="petal_length"
p.yaxis.axis_label="petal_width"
p.line(flowers['petal_length'],flowers['petal_width'])
show(p)
```

```python
x=[3,4,5,6,7]
y=[4,5,6,7,4]
output_file('line.html')
p=figure(title='line graph')
p.line(x,y)
show(p)
```

```python
x=[3,4,5,6,7]
y=[4,5,6,7,4]
output_file('line.html')
p=figure(title='line graph')
p.scatter(x,y)
show(p)
```

```python
x=[3,4,5,6,7]
y=[4,5,6,7,4]
output_file('line.html')
p=figure(title='line graph')
p.scatter(x,y,size=20,fill_color='red')
show(p)
```

```python
x=[3,4,5,6,7]
y=[4,5,6,7,4]
output_file('line.html')
p=figure(title='line graph')
p.scatter(x,y,size=20,fill_color='blue',legend_label='this is my data point')
show(p)
```
