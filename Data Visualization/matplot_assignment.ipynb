#Matplotlib is a Python data visualization library that allows users to create a wide range of static, animated, and interactive visualizations in Python. It provides a variety of plotting functions and tools to create different types of charts, plots, and graphs for data analysis, exploration, and presentation.
#matplotlib is used because it is a powerful and flexible tool for visualizing data in Python. It is easy to use and has a wide range of capabilities, including line plots, scatter plots, bar plots, histograms, pie charts, and many more. Matplotlib allows users to create high-quality and customizable visualizations for scientific and engineering applications, data exploration, and publication-ready graphics.
#Here are five plots that can be plotted using the Pyplot module of Matplotlib:

#Line plot - A line plot shows the relationship between two variables by plotting a line between the data points.

#Scatter plot - A scatter plot shows the relationship between two variables by plotting individual data points.

#Bar plot - A bar plot shows the relationship between a categorical variable and a numerical variable by plotting bars of different heights.

#Histogram - A histogram shows the distribution of a numerical variable by dividing the data into intervals or "bins" and plotting the frequency of data points in each bin.

#Pie chart - A pie chart shows the relative proportions of different categories in a dataset by dividing a circle into sectors proportional to the sizes of the categories.

```python
# What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
#A scatter plot is a type of plot that displays the relationship between two numerical variables. It uses dots to represent each data point, with one variable on the x-axis and the other variable on the y-axis. Scatter plots can show patterns, trends, and correlations between variables.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y,c='b',alpha=.8)
plt.title("Scatter plot of x and y")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
#The subplot() function in Matplotlib is used to create multiple plots in the same figure. It takes three arguments: the number of rows, the number of columns, and the index of the plot to create.
#The subplot() function is useful because it allows us to compare different plots side by side and see how they relate to each other. We can use it to visualize multiple datasets or to explore different aspects of a single dataset.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define data for each line plot
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with 2 rows and 2 columns of plots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

# Plot data for each line in a separate subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add overall title and axis labels to the entire figure
fig.suptitle('Four Line Plots Using Subplots Function')
fig.text(0.5, 0.04, 'X-axis', ha='center')
fig.text(0.04, 0.5, 'Y-axis', va='center', rotation='vertical')

# Display the plot
plt.show()
```

```python
# What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
import matplotlib.pyplot as plt

# Define the data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.bar(company, profit)
plt.title("Company Profit")
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.barh(company, profit)
plt.title("Company Profit")
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")

# Show the plot
plt.show()
```

```python
# What is a box plot? Why is it used? Using the following data plot a box plot
# A box plot is a type of plot used to visualize the distribution of a dataset. It shows the median, quartiles, and outliers of the data in a compact way. The box represents the middle 50% of the data, with the median indicated by a line inside the box. The "whiskers" extend to the minimum and maximum values that are not outliers, and any points outside the whiskers are considered outliers.
#Box plots are useful for identifying skewness, outliers, and overall spread of a dataset. They are often used in statistical analysis to compare the distributions of multiple datasets.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]

# Create a box plot
plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Data")
plt.ylabel("Values")

# Show the plot
plt.show()
```
