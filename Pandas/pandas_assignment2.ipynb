```python
import pandas as pd

def reindex_dataframe(df):
    """
    Re-indexes the given Pandas DataFrame with a new index that starts from 1 and increments by 2 for each row.
    
    Parameters:
    df (Pandas DataFrame): The DataFrame to be re-indexed.
    
    Returns:
    The re-indexed Pandas DataFrame.
    """
    # Create a new index that starts from 1 and increments by 2 for each row
    new_index = pd.RangeIndex(start=1, stop=2*len(df), step=2)
    
    # Set the new index for the DataFrame
    df.set_index(new_index, inplace=True)
    
    # Return the re-indexed DataFrame
    return df
```

```python
# Re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row
df = reindex_dataframe(df)
```

Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the
DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
import pandas as pd
def reindex_df(df):
    return df.reset_index(drop=True).reset_index().apply(lambda row: row * 2 + 1).set_index('index')
```

```python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
print(df)
```

```python
df_reindexed = reindex_df(df)
print(df_reindexed)
```

You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that
iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The
function should print the sum to the console.
For example, if the 'Values' column of df contains the values [10, 20, 30, 40, 50], your function should
calculate and print the sum of the first three values, which is 60.

```python
import pandas as pd
def sum_first_three(df):
    sum = 0
    for i, row in df.iterrows():
        if i < 3:
            sum += row['Values']
    print(sum)
```

```python
df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})
sum_first_three(df)
```

Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column
'Word_Count' that contains the number of words in each row of the 'Text' column.

```python
import pandas as pd
def add_word_count(df):
    df['Word_Count'] = df['Text'].apply(lambda x: len(str(x).split(" ")))
    return df

df = pd.DataFrame({'Text': ['This is a sentence', 'Another sentence with more words', 'And a third sentence']})
df = add_word_count(df)
print(df)
```

How are DataFrame.size() and DataFrame.shape() different?

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
print(df.size)
print(df.shape)
```

As you can see, DataFrame.size() returns a single integer value (9) representing the number of elements in the DataFrame (3 rows x 3 columns). On the other hand, DataFrame.shape() returns a tuple (3, 3) representing the number of rows and columns in the DataFrame, respectively.

You have a Pandas DataFrame df that contains a column named 'Email' that contains email
addresses in the format 'username@domain.com'. Write a Python function that creates a new column
'Username' in df that contains only the username part of each email address.
The username is the part of the email address that appears before the '@' symbol. For example, if the
email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your
function should extract the username from each email address and store it in the new 'Username'
column.

```python
import pandas as pd
def extract_username(df):
    df['Username'] = df['Email'].apply(lambda x: x.split('@')[0])
    return df
df = pd.DataFrame({'Email': ['john.doe@example.com', 'jane.doe@example.com', 'foo.bar@example.com']})
df = extract_username(df)
print(df)

```

The resulting username is assigned to a new column named 'Username' in the DataFrame using the square bracket notation.
Finally, the function returns the modified DataFrame

Here's a Python function that selects all rows from a Pandas DataFrame df where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10:
The function selects the desired rows by indexing the DataFrame with a boolean mask created using the & operator to combine two conditions: one for column 'A' and another for column 'B'.
The function returns a new DataFrame containing only the selected rows.

```python
import pandas as pd

def select_rows(df):
    return df[(df['A'] > 5) & (df['B'] < 10)]
df = pd.DataFrame({'A': [3, 8, 6, 2, 9], 'B': [5, 2, 9, 3, 1], 'C': [1, 7, 4, 5, 2]})
selected_df = select_rows(df)
print(selected_df)
```

Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean,
median, and standard deviation of the values in the 'Values' column.

Here's a Python function that calculates the mean, median, and standard deviation of the values in a Pandas DataFrame df with a column named 'Values':
The function selects the 'Values' column from the DataFrame and calculates the mean, median, and standard deviation using the corresponding methods from Pandas.
The function returns a tuple containing the mean, median, and standard deviation.

```python
import pandas as pd
def calculate_statistics(df):
    values = df['Values']
    mean = values.mean()
    median = values.median()
    std = values.std()
    return mean, median, std
df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})
mean, median, std = calculate_statistics(df)
print(f"Mean: {mean}, Median: {median}, Standard Deviation: {std}")

```

Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python
function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

You can use the Pandas between() function to select all rows where the date is between '2023-01-01' and '2023-01-31'. Here's an example Python function that does this:
The pd.to_datetime() function is used to convert the 'Date' column to a pandas datetime object. Then, the between() function is used to select all rows where the date is between '2023-01-01' and '2023-01-31'. Finally, the function returns the selected data as a new DataFrame.



```python
import pandas as pd

def select_january_data(df):
    # Convert 'Date' column to pandas datetime
    df['Date'] = pd.to_datetime(df['Date'])

    # Select rows where date is between '2023-01-01' and '2023-01-31'
    selected_data = df[df['Date'].between('2023-01-01', '2023-01-31')]

    return selected_data
```
