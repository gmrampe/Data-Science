```python
data1=pd.DataFrame({'key1':[1,2,4,5,6],
                    'key2':[4,5,6,7,8],
                    'key3':[3,4,5,6,6]
})
```

```python
data2=pd.DataFrame({'key1':[1,2,45,6,67],
                    'key4':[56,5,6,7,8],
                    'key5':[3,56,5,6,6]
})
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2)## merge opertion is similar to inner join and return the common elemnts in the given data sets
#The code pd.merge(data1, data2) will merge two Pandas DataFrames data1 and data2 based on their common columns.
#By default, pd.merge() will try to join the two DataFrames on columns that have the same name in both DataFrames. If the columns have different names, or if you want to join on a different column, you can specify the column names to join on using the on argument.
```

```python
pd.merge(data1,data2,how='left')##a match elements return and not matching elements are return nan values 
#The code pd.merge(data1, data2, how='left') will merge two Pandas DataFrames data1 and data2 based on their common columns, using a "left" join.
#A "left" join means that all the rows from data1 are included in the merged DataFrame, along with any matching rows from data2. If there are no matching rows in data2, the corresponding columns in the merged DataFrame will contain NaN values.
```

```python
pd.merge(data1,data2,how='right')
#right is an argument that can be used in the pd.merge() function to specify that the resulting DataFrame should include all the rows from the "right" DataFrame, along with any matching rows from the "left" DataFrame.
#By default, pd.merge() performs an "inner" join, which means that it only includes rows in the resulting DataFrame that have matching values in both DataFrames. However, by specifying right as the how argument, you can include all the rows from the "right" DataFrame in the merged DataFrame, even if there are no matching values in the "left" DataFrame.
```

```python
pd.merge(data1,data2,how='outer')
#how='outer' is another argument that can be used in the pd.merge() function to specify that the resulting DataFrame should include all the rows from both DataFrames, with NaN values in any columns where there are no matching values.
#An "outer" joincombines the rows from both DataFrames based on the values in the specified columns, and includes all the rows from both DataFrames in the resulting DataFrame. If there are no matching values for a row in one of the DataFrames, NaN values are used for the missing columns in the resulting DataFrame.
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2,how='cross')
#pd.merge(data1, data2, how='cross') is another way to perform a cross join between two DataFrames.
#In pandas, the merge() method can be used to combine two DataFrames based on their columns. When how='cross' argument is passed to merge(), it will perform a cross join.
```

```python
pd.merge(data1,data2,how='left',left_on='key2',right_on='key4')
#pd.merge(data1, data2, how='left', left_on='key2', right_on='key4') is a way to perform a left join between data1 and data2 based on different column names.
#In this case, the left_on argument specifies that the join should be performed on the 'key2' column of data1, while the right_on argument specifies that the join should be performed on the 'key4' column of data2.
#A left join returns all the rows from the left DataFrame (data1) and matching rows from the right DataFrame (data2). If there is no match in the right DataFrame, then the values in the resulting DataFrame will be NaN.
```
